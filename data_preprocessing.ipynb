```python
import pandas as pd
pd.options.display.max_columns = 200
pd.options.display.max_rows = 4000
```

```python
# Source dataset: https://data.overheid.nl/dataset/9841-verkeersongevallen---bestand-geregistreerde-ongevallen-nederland#panel-description
# Download: Verkeersongevallen - Bestand geRegistreerde Ongevallen Nederland 2022 - download
# Original filename: ongevallen.txt

df = pd.read_csv('raw_data.txt', sep=",")
```

```python
# Select columns with sufficient amount of values, drop rows with NA

columns = ['ANTL_PTJ', 'AOL_ID', 'NIVEAUKOP', 'GME_ID', 'PVE_CODE', 'WSE_ID', 'BEBKOM', 'WVL_ID', 'WVG_ID', 'WDK_ID', 'LGD_ID', 'WGD_CODE_1', 'AP3_CODE']
df = df[columns].dropna()
df
```

```python
# Anonymize column names

rename_columns = {'ANTL_PTJ': 'x1', 'AOL_ID': 'x2', 'NIVEAUKOP': 'x3', 'GME_ID': 'x4', 'PVE_CODE': 'x5', 'WSE_ID': 'x6', 'BEBKOM': 'x7', 'WVL_ID': 'x8', 'WVG_ID': 'x9', 'WDK_ID': 'x10', 'LGD_ID': 'x11', 'WGD_CODE_1': 'x12', 'AP3_CODE': 'y'}
df = df.rename(rename_columns, axis=1)
```

```python
# Make all columns categorical and map to indexes

df = df.apply(lambda x: x.astype("category").cat.codes)
df
```

```python
# Store dataset

df.to_csv('preprocessed_data.csv', index=True, index_label="index")
```
